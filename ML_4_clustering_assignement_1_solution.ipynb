```python
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p = scipy.misc.face(gray=True)
plt.figure(figsize=(10, 3.6))
plt.imshow(p, cmap=plt.cm.gray)
plt.show()
```

```python
p.shape
```

```python
rows = p.shape[0]
cols = p.shape[1]


image = p.reshape(rows*cols,1)
kmeans = KMeans(n_clusters = 5)
kmeans.fit(image)

clusters = np.asarray(kmeans.cluster_centers_) 
labels = np.asarray(kmeans.labels_)  
labels = labels.reshape(rows,cols); 
```

```python
plt.imsave('rocket.png',labels);
image = plt.imread('rocket.png')
plt.figure(figsize=(10, 3.6))
plt.imshow(image)
plt.show()
```
